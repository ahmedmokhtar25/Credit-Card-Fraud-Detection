### We will use the isolationforest algorithm to detect anomalies

```python
# check the scale of the dataframe
df_describe = pd.DataFrame(credit_card.describe())
df_describe
```

```python
# split the data into independent and dependent variables
x = credit_card.drop(columns= 'Class')
y= credit_card[['Class']]
```

```python
# run Isolation forest model
from sklearn.ensemble import IsolationForest
is_forest = IsolationForest(n_estimators= 300, max_samples= 100000, contamination = 0.2, bootstrap= True, random_state = 0)
otliers_prediction = is_forest.fit_predict(x)
otliers_prediction
```

```python
# get the indices of the outliers 
no_of_outliers = []
for i in range(len(otliers_prediction)):
    if otliers_prediction[i] == -1:
        no_of_outliers.append(i)
len(no_of_outliers), no_of_outliers
```

```python
# anomaly score
anomaly_score = (is_forest.decision_function(x))
anomaly_score
```

```python
# set a threshold 10%
anomaly_score = -1*anomaly_score
percentile_threshold = np.percentile(anomaly_score, 90)
percentile_threshold
```

```python
# disply the anomaly score
anomaly_score
```

```python
# filter the anomalies by the threshold
outlier_threshold = np.where(anomaly_score>= percentile_threshold)[0]
outlier_threshold
```

```python
# check the number of the outliers
len(outlier_threshold)
```

```python
# drop the outliers
banking_version_one = credit_card.copy(deep= True)
banking_version_one.drop(index= outlier_threshold, inplace= True)
```

```python
# Check the outliers
fig, axes = plt.subplots(4,8,figsize = (20,20))
axes = axes.flatten()
for i, column in enumerate(banking_version_one.columns):
    sns.boxplot(banking_version_one[column], ax= axes[i])
    axes[i].set_title(f'Box Plot for {column}')

plt.tight_layout()
plt.show()
```

```python
# check the imbalance
banking_version_one['Class'].value_counts()
```

```python
# divide the data into independent and dependent variables
x=  banking_version_one.drop(columns= 'Class')
y= banking_version_one[['Class']]
```

```python
len(x), len(y)
```

```python
# splitting data to train and test 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

## Deal with the strong correlations between independent variables

```python
# using PCA to transform correlated variables to uncorrelated compoenents
from sklearn.decomposition import PCA
pca = PCA(random_state= 0)
pca.fit(x_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >=0.95) + 1
d
```

```python
# display the number of dimensions and the eigenvalues 
plt.figure(figsize= (10,5))
sns.lineplot(x = np.arange(1,30,1), y = cumsum)
plt.xticks(np.arange(1,30,2))
plt.axhline(0.95, linestyle = '-', color ='r')
plt.show()
```

## The best number of dimensions is 22

```python
# get the new dimensions
pca = PCA(n_components=22, random_state=0)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
# implement a logisitic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score   , precision_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
lg_model = LogisticRegression(random_state= 0)
parameters = {'penalty': ['l1', 'l2'], 'C': [1, 5, 10], 'solver' : ['saga', 'lbfgs']}
lg_grid_model = GridSearchCV(estimator= lg_model, param_grid = parameters, cv = 5, scoring= 'recall')
lg_grid_model.fit(x_train_pca,y_train)
```

```python
# the outcomes of the logistic regression
lg_grid_model.best_score_, lg_grid_model.best_params_, lg_grid_model.cv_results_
```

```python
# implement the confusion matrix of the logisitic regression
from sklearn.metrics import confusion_matrix
y_pred = lg_grid_model.best_estimator_.predict(x_test_pca)
plt.figure(figsize= (10,5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot= True, fmt= '.5g')
```

```python
logistic_score = recall_score(y_test, y_pred)
```

```python
# implement a random forest model
from sklearn.ensemble import RandomForestClassifier
rn_model = RandomForestClassifier(random_state=0, bootstrap= True)
parameters = {'criterion': ['entropy', 'gini'], 'max_depth': [6,9,10], 'max_samples' : [18000,20000]}
rn_grid_model = GridSearchCV(estimator= rn_model, param_grid = parameters, cv = 5, scoring= 'recall')
rn_grid_model.fit(x_train_pca,y_train)
```

```python
# Random forest model outcomes
rn_grid_model.best_score_, rn_grid_model.best_params_
```

```python
y_pred = rn_grid_model.best_estimator_.predict(x_test_pca)
sns.heatmap(confusion_matrix(y_test, y_pred), annot= True, fmt= '.5g')
```

```python
RandomForest_score = recall_score(y_test, y_pred)
```

```python
logistic_score, RandomForest_score
```
